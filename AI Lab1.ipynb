```python
lst = list(map(int,input("Enter the list using space: ").split()))
print("You have entered ", len(lst), "items")
print("Print the 3th item: ", lst[2])
```

```python
lst[2] = 100
print("Print the 3th item after change: ", lst[2])
```

```python
lst.append(512)
print("Add an item in the list : ", lst)
```

```python
lst1 = sorted(lst)
print("Sort the list: ", lst1)
```

```python
lst.sort(reverse = True)
print("Sorted reverse the list: ", lst)
```

```python
print("Slice the list from 2th to 5th: ", lst[2:6])
```

```python
print("Print the list from the last using slice: ", lst[::-1])
```

```python
lst2 = list(map(int,input("Enter the 2nd list using space: ").split()))

print("Adding two lst: ", lst+lst2)
```
